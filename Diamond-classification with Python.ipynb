```python
x = data.drop(columns=['Unnamed: 0','cut','color','clarity','price']).to_numpy()
#x=pd.DataFrame(np.c_[data['age'],data['bmi'],data['children'],columns = ['age','bmi','children'])
#x = data[['age', 'bmi', 'children']]
x
```

```python
y=data['price']
y
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y,"r+")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
linreg1=LinearRegression()
```

```python
# Entrainer le modèle avec la méthode fit
linreg1.fit(X_train,y_train)
linreg1.score(X_train,y_train)
```

```python
a0=linreg1.intercept_
a0
```

```python
[a1,a2,a3,a4,a5,a6]=linreg1.coef_
print(a1,a2,a3,a4,a5,a6)
```

```python
t=np.linspace(0,30,10)
plt.plot(x,y,'ro')
plt.plot(t,a0+a1*t+a2*t+a3*t+a4*t+a5*t+a6*t)
```
